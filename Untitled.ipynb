```python
import pandas as pd
import numpy as np


dict={"first_score":[100,23,34,np.nan],
     "second_score":[np.nan,np.nan,4,12],
     "third_score":[22,26,np.nan,34]}
 
# creating a dataframe from list
dict=pd.DataFrame(dict)
print(dict)
print("\n \n")
# using isnull() function 
dict.isnull()
#filling missing value using fillna()
print(dict.fillna(0))

print("\n \n")
# filling a missing value with
# previous ones
print(dict.fillna(method='pad'))


# filling  null value using fillna() function
dict.fillna(method="bfill")
```

```python
# using isnull() function
dict.notnull()
```

```python
import pandas as pd
import numpy as np


dict={"first_score":[np.nan,23,34,np.nan],
     "second_score":[np.nan,np.nan,4,12],
     "third_score":[np.nan,26,np.nan,34]}
 
# creating a dataframe from list
dict=pd.DataFrame(dict)
print(dict)
#suppresion des ligne qui possede tous les valeurs une valeur null
dict.dropna(how='all')
```

```python
# suppresion des colonnes qui possede  au moins une valeur null
dict.dropna(axis=1)
```

```python
# making data frame from csv file
data = pd.read_csv("employees.csv")
new_data=data.dropna(axis=0,how="any")

print("\n Old data frame length:", len(data))
print("\n New data frame length:",len(new_data))

print("\nNumber of rows with at least 1 NA value: ", len(data)-len(new_data))
```
